```python
depart = "2024-05-02T07:00:00"
arrival = "2024-05-02T18:00:00"
origin = "25.047743,121.516273"
destination = "24.802340,120.970486"
route_return = 3
transit = "3,4,5,6,7,8,9"
getkey.getmass(f"https://tdx.transportdata.tw/api/maas/routing?origin={origin}&destination={destination}&gc=0.0&top={route_return}&transit={transit}&transfer_time=15%2C60&depart={depart}&arrival={arrival}&first_mile_mode=0&first_mile_time=10&last_mile_mode=0&last_mile_time=10")
```

```python
with open('Attractions.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
scenic = pd.DataFrame(data)
scenic
```

[POPUP 作法](https://python-visualization.github.io/folium/latest/user_guide/ui_elements/popups.html)

```python
#getkey.Updatescenic_map
Attraction = []

Taiwan_map = folium.Map(location=[23.800983, 120.772569], tiles='openstreetmap', zoom_start=7.3)

for row in scenic.iloc:
    Attraction = (row['Position']['PositionLat'], row['Position']['PositionLon'], row['ActivityName'], row['Description'])
    html = f"""
    <div style="width:400px;height:400px">
        <h1> {Attraction[2]}</h1><br>
    
    <p>
    {Attraction[3][:20] + "..."}
    </p>
    </div>

    """
    mm = folium.Marker(location=(Attraction[0], Attraction[1]),#座標
                       popup=html, #車牌
    )
    Taiwan_map.add_child(mm)

Taiwan_map
```

```python
Taiwan_map.save('taiwan.html')
```

```python

Taiwan_map = folium.Map(location=[23.800983, 120.772569], tiles='openstreetmap', zoom_start=7.3)
color = ['red', 'blue', 'orange', 'green']
index = 0
for i in glob.glob('scenic_*'):

    #popup = folium.GeoJsonPopup(fields=[f"Popup {index}"])

    folium.GeoJson(
        i,
        style_function=lambda feature: {
        "fillColor": "red" if index%2 == 0 else "green",
        "color": "black",
        "weight": 1,
        "dashArray": "5, 5",
        "fillOpacity": 0.2
        },

    ).add_to(Taiwan_map)
    
    index = (index + 1)%len(color)
```
