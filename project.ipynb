```python
import getkey
import json
import pandas as pd
import folium
import glob

```

```python
getkey.getjson('https://tdx.transportdata.tw/api/basic/v2/Tourism/ScenicSpot?%24top=50&%24format=JSON', "scenicSpot")
getkey.getjson('https://tdx.transportdata.tw/api/basic/v2/Tourism/Activity?%24top=50&%24format=JSON', "Attractions_activity")
getkey.getjson('https://tdx.transportdata.tw/api/tourism/service/odata/V2/Tourism/Attraction?%24top=50', "Attractions")
```

```python
with open('data/Attractions.json', 'r', encoding='utf-8') as f:
    Attractions = json.load(f)['value']
```

```python
with open('data/Attractions_activity.json', 'r', encoding='utf-8') as f:
    Attractions_activity = json.load(f)
```

```python
with open('data/scenicSpot.json', 'r', encoding='utf-8') as f:
    scenicSpot = json.load(f)
```

```python
pd_scenicSpot = pd.DataFrame(scenicSpot)
pd_scenicSpot.head()
```

```python
pd_Attractions = pd.DataFrame(Attractions)
pd_Attractions.head()
```

```python
pd_Attractions_activity = pd.DataFrame(Attractions_activity)
pd_Attractions_activity.head()
```

```python
pd_scenicSpot = pd.DataFrame(scenicSpot)
pd_scenicSpot=pd_scenicSpot.fillna(0)
pd_scenicSpot.tail()
```

[POPUP 作法](https://python-visualization.github.io/folium/latest/user_guide/ui_elements/popups.html)

[點擊活動](https://stackoverflow.com/questions/74707544/add-a-clickevent-function-to-multiple-folium-markers-with-python)

```python


Taiwan_map = folium.Map(location=[23.800983, 120.772569], tiles='openstreetmap', zoom_start=7.3)

for row in pd_Attractions_activity.iloc:
    Attraction = (row['Position']['PositionLat'], row['Position']['PositionLon'], row['ActivityName'], row['Description'])
    html = f"""
    <div style="width:400px;height:300px">
        <h1> {Attraction[2]}</h1><br>
        <p>
        {Attraction[3][:100] + "..."}
        </p>
    </div>
    """
    mm = folium.Marker(location=(Attraction[0], Attraction[1]),#座標
                       popup=html, #車牌
                       icon=folium.Icon(color='orange', icon='info-sign'),
    )
    Taiwan_map.add_child(mm)

for row in pd_Attractions.iloc:
    Attraction = (row['PositionLat'], row['PositionLon'], row['AttractionName'], row['Description'])
    html = f"""
    <div style="width:400px;height:200px">
        <h1> {Attraction[2]}</h1><br>
    
    <p>
    {Attraction[3][:100] + "..."}
    </p>
    </div>

    """
    mm = folium.Marker(location=(Attraction[0], Attraction[1]),#座標
                       popup=html, #車牌
                       icon=folium.Icon(color='blue', icon='info-sign'),
    )
    Taiwan_map.add_child(mm)

for row in pd_scenicSpot.iloc:
    Attraction = (row['Position']['PositionLat'], row['Position']['PositionLon'], row['ScenicSpotName'], row['DescriptionDetail'])
    html = f"""
    <div style="width:400px;height:200px;background='green'">
        <h1> {Attraction[2]}</h1><br>
    
    <p>
    {Attraction[3][:100] + "..."}
    </p>
    </div>

    """
    mm = folium.Marker(location=(Attraction[0], Attraction[1]),#座標
                       popup=html, #車牌
                       icon=folium.Icon(color='green', icon='info-sign'),
    )
    Taiwan_map.add_child(mm)

Taiwan_map
```

```python
Taiwan_map.save('taiwan.html')
```
